```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("./PROJECT_IPL_1_Edited.csv")
data.rename(columns={'Player Name':'Player_Name','Base Price (in Lakhs Rupees)':'Base_Price','Price Paid (in Lakhs Rupees)':'Price_Paid','Status':'Status','Bought by':'Bought_by','Previous IPL Teams':'Previous_IPL_Teams','Country':'Country','Role':'Role','Batting Style':'Batting_Style','Bowling Style':'Bowling_Style','Batting Order':'Batting_Order','Capped or Uncapped at Internationals (during Auction)':'Capped_or_Uncapped'},inplace='True')
data.loc[data["Role"] == "Fast Bowler", "Role"] = "Seam Bowler"
```

<h2 style="text-align:center;"> Listing out players bought by RCB </h2>

```python
rcb=data.groupby('Bought_by').get_group('RCB')
rcb
```

<h2 style="text-align:center;"> Visualization of Price Paid for players bought by RCB </h2>

```python
x=list(rcb['Player_Name']);
y=list(rcb['Price_Paid']);
plt.barh(x,y);
```

<h2 style="text-align:center;">List of Batsmen bought by RCB</h2>

```python
rcb_bats=rcb[rcb['Role'].str.contains('Batsman')]
rcb_bats
```

<h2 style="text-align:center;">List of Bowlers bought by RCB</h2>

```python
rcb_bowl=rcb[rcb['Role'].str.contains('Bowler')]
rcb_bowl
```

<h2 style="text-align:center;">List of All Rounders bought by RCB</h2>

```python
rcb_all=rcb[rcb['Role'].str.contains('All Rounder')]
rcb_all
```

<h2 style="text-align:center;">Analysis - Royal Challengers Bangalore </h2>

<h3>Total amount spent during auction (in crores): </h3>

```python
rcb['Price_Paid'].sum()/100
```

<h3>Highest bid by Royal Challengers Bangalore: </h3>

```python
rcb[rcb['Price_Paid']==rcb['Price_Paid'].max()]
```

<h3>Highest Indian bid by Royal Challengers Bangalore: </h3>

```python
rcb[rcb['Price_Paid']==rcb[rcb['Country']=='India']['Price_Paid'].max()]
```

<h3>Highest Foreign bid by Royal Challengers Bangalore: </h3>

```python
rcb[rcb['Price_Paid']==rcb[rcb['Country']!='India']['Price_Paid'].max()]
```

<h3>Amount Spent on each Department (in crores): </h3>

```python
print((rcb.groupby('Role')['Price_Paid'].sum()/100).to_string())
```

<h3>Highest batsman bid by Royal Challengers Bangalore: </h3>

```python
rcb_bats[rcb_bats['Price_Paid']==rcb_bats['Price_Paid'].max()]
```

<h3>Highest bowler bid by Royal Challengers Bangalore: </h3>

```python
rcb_bowl[rcb_bowl['Price_Paid']==rcb_bowl['Price_Paid'].max()]
```

<h3>Highest All Rounder bid by Royal Challengers Bangalore: </h3>

```python
rcb_all[rcb_all['Price_Paid']==rcb_all['Price_Paid'].max()]
```

<h3>Highest Capped bid by Royal Challengers Bangalore: </h3>

```python
rcb[rcb['Price_Paid']==rcb[rcb['Capped_or_Uncapped']=='Capped']['Price_Paid'].max()]
```

<h3>Highest Uncapped bid by Royal Challengers Bangalore: </h3>

```python
rcb[rcb['Price_Paid']==rcb[rcb['Capped_or_Uncapped']=='Uncapped']['Price_Paid'].max()]
```

<h3>Visualization of distribution of Amount spent by RCB on each department: </h3>

```python
rcb.groupby(['Role']).sum().plot(kind='pie', y='Price_Paid', autopct="%.2f%%", figsize=(15,10), title="Distribution of Amount spent by RCB on each department").legend(loc='upper right');
```

<h3>No. of IPL Experienced and inexperienced Players bought by Royal Challengers Bangalore: </h3>

```python
print("Experienced Players:     ",rcb[pd.notnull(rcb['Previous_IPL_Teams'])].shape[0])
print("Inexperienced Players:   ",rcb[pd.isnull(rcb['Previous_IPL_Teams'])].shape[0])
```
